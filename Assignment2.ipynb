```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the dataset

```python
dataset=pd.read_csv('/content/House Price India.csv')
```

Univariate Analysis

```python
dataset.shape
```

```python
column_names=dataset.columns.tolist()
print(column_names)
```

```python
for i in range(0,23):
  plt.figure(figsize=(5,3))
  sns.histplot(dataset[column_names[i]],kde=True)
  plt.xlabel(column_names[i])
  plt.ylabel('Frequency')
  plt.ylabel('Histogram for {}'.format(column_names[i]))
  plt.show()
```

```python
for i in range(0,23):
  plt.figure(figsize=(5,3))
  sns.boxplot(x=dataset[column_names[i]])
  plt.xlabel(column_names[i])
  plt.title("Box plot for {}".format(column_names[i]))
  plt.show()
```

```python
for i in range(0,23):
  mean_value=dataset[column_names[i]].mean()
  median_value=dataset[column_names[i]].median()
  mode_value=dataset[column_names[i]].mode()[0]
  variance_value=dataset[column_names[i]].var()
  std_deviation=dataset[column_names[i]].std()
  print(column_names[i].upper())
  print("Mean:{}, Median:{}, Mode:{}, Variance:{}, Standard Deviation:{}".format(mean_value,median_value,mode_value,variance_value,std_deviation))
```

Bivariate Analysis

```python
for i in range(1,23):
  a=dataset[column_names[i-1]]
  b=dataset[column_names[i]]
  plt.scatter(a,b)
  plt.xlabel(column_names[i-1])
  plt.ylabel(column_names[i])
  plt.show()
```

```python
for i in range(1,23):
  a=dataset[column_names[i-1]]
  b=dataset[column_names[i]]
  corrcoeff=a.corr(b)
  print("Coefficient of correlation for {} wrt {}".format(column_names[i-1],column_names[i]))
  print(corrcoeff)
```

Multivariate Analysis

```python
correlation_matrix=dataset.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
