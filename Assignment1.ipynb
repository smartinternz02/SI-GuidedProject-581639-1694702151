1)Creating a dataframe df

```python
import pandas as pd
import numpy as np
```

```python
data=np.random.rand(4,4)
```

```python
features=['Feature1','Feature2','Feature3','Feature4']
df=pd.DataFrame(data,columns=features)
```

```python
df
```

2) Renaming the columns

```python
new_names=['Random value 1','Random value 2','Random value 3','Random value 4']
df.rename(columns=dict(zip(features,new_names)),inplace=True)
df
```

3) Descriptive statistics

```python
df.describe()
```

4) Checking for null values and printing the datatypes of columns

```python
df.isnull()
```
